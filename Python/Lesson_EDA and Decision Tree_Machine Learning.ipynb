```python
features = data.iloc[:,:-1].values
label = data.iloc[:,[-1]].values
```

```python
features
```

```python
#Libraries functions to convert into 0,1

#Convert categorical features to numerical features as sklearn works only with mupy array
#Instead of label enconding and then onehotencoding, 
#new version directly works with onehotencoding using ColumnTransformer

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
transformer = ColumnTransformer(transformers=[
        ("OneHot", #Just a name
         OneHotEncoder(),   #the transfomer flass
         [3]     #the column to be applied on
        )
    ],
    remainder='passthrough' #do not apply anything to the remaining colimns
)

features = transformer.fit_transform(features.tolist())   #to convert to list
features
```

```python
#convert an object to normal array

features = features.astype(float)
```

```python
features
```

# Apply ML Algorithm

```python
from sklearn.model_selection import train_test_split
```

```python
X = features
Y = label


X_train, X_test, Y_train, Y_test = train_test_split(features, label, test_size = 0.2, random_state=1)
```

```python
#Create model using DT Reg

from sklearn.tree import DecisionTreeRegressor
DTR = DecisionTreeRegressor(max_depth = 3)
DTR.fit(X_train, Y_train)

```

```python
#Checking score of training as well as testing

print(DTR.score(X_train, Y_train))
print(DTR.score(X_test, Y_test))
```
